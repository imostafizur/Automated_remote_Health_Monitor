### ECG csv dataset heartbeat abnormality detection

```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

#### Normal heart beat

```python
# Load csv dataset
df = pd.read_csv('ptbdb_normal.csv')
```

```python
# dataset details
print(df.shape)
```

```python
# dataset info
print(df.info())
```

```python
# Abnoraml hearbeat csv load
df_abnormal = pd.read_csv('ptbdb_abnormal.csv')
```

```python
# Join two dataset and label to normal and abnormal
df = pd.concat([df, df_abnormal], axis=0) 
```

```python
df.head()
```

```python
# number of features
print(df.shape)
```

```python
# Ecg signal plot
plt.figure(figsize=(50, 5))
plt.plot(df.iloc[0, :])
plt.title('ECG Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# ECG signal plot using seaborn
plt.figure(figsize=(20, 5))
sns.lineplot(data=df.iloc[0, :])
plt.title('ECG Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```
